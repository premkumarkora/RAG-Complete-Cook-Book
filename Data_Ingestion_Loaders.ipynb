```python
# Example: Loading a PDF
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("/Volumes/vibecoding/RAG-Complete Cook Book/testing PDF.pdf")
docs = loader.load()
# Now 'docs' contains the text from the PDF
```

We have converted the PDF to text and stored it in a variable called docs.

```python
print(docs)
```
